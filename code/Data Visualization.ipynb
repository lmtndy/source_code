```python
from pymongo import MongoClient

# Thay đổi các giá trị dưới đây theo thông tin của bạn
mongo_host = 'localhost'
mongo_port = 27017
mongo_db = 'DIEMTHITHPT'

# Tạo đối tượng MongoClient
client = MongoClient(host=mongo_host, port=mongo_port)

# Chọn database
db = client[mongo_db]

# In danh sách các collection trong database
print(db.list_collection_names())
```

```python
collection_name = 'diem_thi'
collection = db[collection_name]
```

```python
df = pd.DataFrame(list(collection.find()))
```

```python
df.head()
```

```python
df.describe()
```

```python
df[df['gioi_tinh'] == 'Nữ']
```

```python
filter_criteria = {'gioi_tinh': 'Nữ'}

# Find documents that match the filter criteria
filtered_documents = list(collection.find(filter_criteria))

# Convert the filtered MongoDB collection to a DataFrame
df_filtered = pd.DataFrame(filtered_documents)
# Display the DataFrame
print("Filtered DataFrame:")
print(df_filtered)
```
