```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pickle
import warnings
warnings.filterwarnings("ignore")
```

```python
# config
plt.rcParams["figure.figsize"] = (10, 8)
plt.rcParams["figure.dpi"] = 100
plt.rcParams["font.size"] = 14
```

```python
# load data
df = pd.read_csv("D:/Study/Truc_quan_hoa_data/Thuc_hanh/data/Income.csv")
x = df[["Income"]]
y = df[["Expenditure"]]
```

```python
# fit model
model = LinearRegression().fit(x, y)
```

```python
# get result
intercept = model.intercept_
slope = model.coef_
R_2 = model.score(x, y)
```

```python
# visualization
plt.scatter(x, y)
plt.plot(x, intercept + slope * x)
plt.xlabel("Income")
plt.ylabel("Expenditure")
plt.show()
```

```python
# prediction
# pre_value = intercept + slope * x
pre_values = model.predict(x)
plt.plot(x, y, label="Actual", color="red", linestyle="--")
plt.plot(x, pre_values, label="Predict")
plt.legend()
plt.show()
```

```python
# save model
with open('models/linearRegression', 'wb') as f:
    pickle.dump(model, f)
```
